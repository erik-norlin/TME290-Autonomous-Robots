```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
t = []
mean = []
max = []
line = 0
with open('../status.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        if line > 0:
            t.append(float(row[0]))
            mean.append(float(row[1]))
            max.append(float(row[2]))
        line += 1
print(np.mean(mean[100:]))
```

```python
fontsize = 14
figsize = 8
step = 1
fig, ax = plt.subplots(figsize=(figsize,figsize))
ax.plot(t[:-step], mean[:-step], label='Mean')
ax.plot(t[:-step], max[:-step], label='Max')
# ax.set_xticks([0,0.5,1])
ax.set_yticks(np.linspace(0.4,1,13))
ax.set_xlim([t[0], t[-step]])
ax.set_ylim([0.4, 1.03])
ax.set_xlabel('$t$ [min]', fontsize=fontsize)
ax.set_ylabel('Amount of grass', fontsize=fontsize)
ax.legend(loc='center right', fontsize=fontsize)
plt.savefig('status_plot6.png')
plt.show()
print(min(mean))
```
